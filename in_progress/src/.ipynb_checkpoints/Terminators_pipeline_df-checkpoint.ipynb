## Derectly search with no mismatches/indels allowed 

````python
# Functions
def compl_letter(letter):
    if letter == 'A':
        return 'T'
    elif letter == 'C':
        return 'G'
    elif letter == 'G':
        return 'C'
    elif letter == 'T':
        return 'A'
    elif letter == '-':
        return '-'
    elif letter == '@':
        return '@'
    elif letter == 'N':
        return 'N'
    else:
        raise NameError, 'Unknown letter: ' + letter

def compl(seq):
    seq = seq.upper()
    compl_seq = ''
    for i in range(len(seq)-1, -1, -1):
        compl_seq += compl_letter(seq[i])
    return compl_seq

def search_fasta(pattern, file_path, file_format):
    matches = []
    for record in SeqIO.parse(open(file_path, "rU"), file_format):
        chrom = record.id
        for match in re.finditer(pattern, str(record.seq)):
            # 1-based
            start_pos = match.start() + 1
            end_pos = match.end()
            matches.append((start_pos, end_pos))
    return matches

# Terinators file reader
for i in range(2):
    
    if i == 0: # Ec (by genome)
        print 'E. coli (by genome)'
        processed_term_file_path = Ec_terminators_file_path
        map_terminators_file_path = Ec_map_term_file_path
        curr_term_df = Ec_term_df
        genome_file_path = Ec_genome_file_path
        genome_file_format = Ec_genome_file_format
        BED_terminators_file_path = Ec_BED_terminators_file_path
        BED_first_line = Ec_BED_first_line
        BED_chrom_colname = Ec_BED_chrom_colname
    elif i == 1: # Sp (by genome)
        print 'S. pullorum (by genome)'
        processed_term_file_path = Sp_terminators_file_path
        map_terminators_file_path = Sp_map_term_file_path
        curr_term_df = Sp_term_df
        genome_file_path = Sp_genome_file_path
        genome_file_format = Sp_genome_file_format
        BED_terminators_file_path = Sp_BED_terminators_file_path
        BED_first_line = Sp_BED_first_line
        BED_chrom_colname = Sp_BED_chrom_colname
        
    t_file_reader = open(processed_term_file_path,'r')
    header_array = t_file_reader.readline().split()
    #print header_array 
    seq_col_index = header_array.index(seq_col_name)
    #print seq_col_name
    #print seq_col_index
    org_col_index = header_array.index(org_col_name)
    score_col_index = header_array.index(score_col_name)

    # Mapped terminators file writer
    out_writer = open(map_terminators_file_path, 'wb')
    tsv_writer = csv.writer(out_writer, delimiter='\t', lineterminator="\n")

    # Hearder of the mapped terminators file
    map_header_array = header_array + ['strand', 'start', 'end']
    tsv_writer.writerow(map_header_array)

    # BED file writer
    BED_writer = open(BED_terminators_file_path, 'wb')
    BED_writer.write(BED_first_line + "\n")
    tsv_BED_writer = csv.writer(BED_writer, delimiter='\t', lineterminator="\n")
    
    for t_line in t_file_reader:
        t_array = t_line.split()
        #print t_array
        dir_pattern = t_array[seq_col_index]
        #print dir_pattern
        dir_coords = search_fasta(dir_pattern, genome_file_path, genome_file_format)

        rev_pattern = compl(dir_pattern)
        #print rev_pattern
        rev_coords = search_fasta(rev_pattern, genome_file_path, genome_file_format)

        if len(dir_coords) + len(rev_coords) > 1:
            TypeError('Several matches')
        if len(dir_coords) + len(rev_coords) == 0:
            strand = 'NA'
            coords = ['NA', 'NA']
        if len(dir_coords) == 1:
            strand = '+'
            coords = dir_coords[0]
        if len(rev_coords) == 1:
            strand = '-'
            coords = rev_coords[0]
        map_array = t_array + [str(strand), coords[0], coords[1]] 
        # Exclude NA from the analysis
        if strand != 'NA':
            tsv_writer.writerow(map_array)
            score = t_array[score_col_index]
            BED_array = [BED_chrom_colname, coords[0]-1, coords[1], t_array[org_col_index], str(score), str(strand)] # 0-based 
            tsv_BED_writer.writerow(BED_array)
         
    t_file_reader.close()
    out_writer.close()
    BED_writer.close()
    print 'File '+ BED_terminators_file_path.split('/')[-1] + ' has been created' 
    print 'File '+ map_terminators_file_path.split('/')[-1] + ' has been created' 

#print 'Mapped .tsv files have been created'
#print 'BED files have been created'
````

## Upstream and downstream genes
* Use gff file to extract closest genes (check for intersections as well): write down their names, ids and coordinates
* Distances between termonator and CDS end (upstream gene), CDS start (downstream gene)
* Join to mapped tsv files and filter out unnecesaary columns

````python
from ppretty import ppretty

Ec_terms = pybedtools.BedTool(Ec_BED_terminators_file_path).sort()
Sp_terms = pybedtools.BedTool(Sp_BED_terminators_file_path).sort()

# Include only 'gene' rows
Ec_genes = pybedtools.BedTool(Ec_annotation_file_path).sort()
Ec_genes = Ec_genes.filter(lambda x: x[2] == 'gene').saveas()

Sp_genes = pybedtools.BedTool(Sp_annotation_file_path).sort()
Sp_genes = Sp_genes.filter(lambda x: x[2] == 'gene').saveas()


# Find upstream genes -> pandas df
EC_up = Ec_terms.closest(Ec_genes, D='a', fu=True, io=False)
EC_up_df = EC_up.to_dataframe(names =['genome', 'term_start', 'term_end', 'term_name', 'score', 'term_strand', 'genome_again', 'ann_based_on', 'feature_type', 'up_gene_start', 'up_gene_end', 'smth', 'up_gene_strand', 'smth2', 'up_gene_descr', 'up_dist'])
#display(EC_up_df.head())

SP_up = Sp_terms.closest(Sp_genes, D='a', fu=True, io=False)
SP_up_df = SP_up.to_dataframe(names =['genome', 'term_start', 'term_end', 'term_name', 'score', 'term_strand', 'genome_again', 'ann_based_on', 'feature_type', 'up_gene_start', 'up_gene_end', 'smth', 'up_gene_strand', 'smth2', 'up_gene_descr', 'up_dist'])
#display(SP_up_df.head())

# Find downstream genes -> pandas df
EC_dn = Ec_terms.closest(Ec_genes, D='a', fd=True, io=False)
EC_dn_df = EC_dn.to_dataframe(names =['genome', 'term_start', 'term_end', 'term_name', 'score', 'term_strand', 'genome_again', 'ann_based_on', 'feature_type', 'dn_gene_start', 'dn_gene_end', 'smth', 'dn_gene_strand', 'smth2', 'dn_gene_descr', 'dn_dist'])
#display(EC_dn_df.head())

SP_dn = Sp_terms.closest(Sp_genes, D='a', fd=True, io=False)
SP_dn_df = SP_dn.to_dataframe(names =['genome', 'term_start', 'term_end', 'term_name', 'score', 'term_strand', 'genome_again', 'ann_based_on', 'feature_type', 'dn_gene_start', 'dn_gene_end', 'smth', 'dn_gene_strand', 'smth2', 'dn_gene_descr', 'dn_dist'])
#display(SP_dn_df.head())


# Make these dataframes cute
## Delete unusual columns
EC_up_df = EC_up_df[['term_name', 'term_start', 'term_end', 'term_strand', 'up_gene_descr', 'up_gene_start', 'up_gene_end', 'up_gene_strand', 'up_dist']]
#display(EC_up_df.head())

SP_up_df = SP_up_df[['term_name', 'term_start', 'term_end', 'term_strand', 'up_gene_descr', 'up_gene_start', 'up_gene_end', 'up_gene_strand', 'up_dist']]
#display(SP_up_df.head())

EC_dn_df = EC_dn_df[['term_name', 'term_start', 'term_end', 'term_strand', 'dn_gene_descr', 'dn_gene_start', 'dn_gene_end', 'dn_gene_strand', 'dn_dist']]
#display(EC_up_df.head())

SP_dn_df = SP_dn_df[['term_name', 'term_start', 'term_end', 'term_strand', 'dn_gene_descr', 'dn_gene_start', 'dn_gene_end', 'dn_gene_strand', 'dn_dist']]
#display(SP_up_df.head())

## Add gene names (strip gene decriptions)
def extract_gene_name(gene_descr, org):
    if org == 'Ec':
        gene_name = gene_descr.strip().replace('Name=', '').replace('gene', '').strip()
    elif org == 'Sp':
        gene_name = gene_descr.split(';')[1].strip().replace('Name=', '').strip()
    else:
        raise NameError, 'Unknown organism: ' + str(org)
    
    return gene_name

EC_up_df['up_gene_name'] = EC_up_df['up_gene_descr'].apply(lambda x: extract_gene_name(x, 'Ec'))
#display(EC_up_df.head())
print EC_up_df.shape

SP_up_df['up_gene_name'] = SP_up_df['up_gene_descr'].apply(lambda x: extract_gene_name(x, 'Sp'))
#display(SP_up_df.head())
        
EC_dn_df['dn_gene_name'] = EC_dn_df['dn_gene_descr'].apply(lambda x: extract_gene_name(x, 'Ec'))
#display(EC_dn_df.head())

SP_dn_df['dn_gene_name'] = SP_dn_df['dn_gene_descr'].apply(lambda x: extract_gene_name(x, 'Sp'))
#display(SP_dn_df.head())


# Join to processed terminators dataframe ()
Ec_term_df_new = Ec_term_df.merge(EC_up_df, left_on='Name', right_on='term_name', how='outer', validate='one_to_one').merge(EC_dn_df, on=['term_name', 'term_start', 'term_end', 'term_strand'], how='outer', validate='one_to_one')
#display(Ec_term_df_new.head())
#print Ec_term_df_new.columns
Sp_term_df_new = Sp_term_df.merge(SP_up_df, left_on='Name', right_on='term_name', how='outer', validate='one_to_one').merge(SP_dn_df, on=['term_name', 'term_start', 'term_end', 'term_strand'], how='outer', validate='one_to_one')
#display(Ec_term_df_new.head())

# Rename columns
Ec_term_df_new.rename(columns = {'Replicate.1_x':'rep1_x', 'Replicate.2_x':'rep2_x','Replicate.3_x':'rep3_x','AVERAGE_x': 'AVG_x', 'STDEV_x':'STD_x', 'Replicate.1_y':'rep1_y', 'Replicate.2_y':'rep2_y','Replicate.3_y':'rep3_y','AVERAGE_y': 'AVG_y', 'STDEV_y':'STD_y', 'Free_energy': 'dG', 'functional_x':'is_funct_x', 'functional_y':'is_funct_y'}, inplace = True)
Sp_term_df_new.rename(columns = {'Replicate.1_x':'rep1_x', 'Replicate.2_x':'rep2_x','Replicate.3_x':'rep3_x','AVERAGE_x': 'AVG_x', 'STDEV_x':'STD_x', 'Replicate.1_y':'rep1_y', 'Replicate.2_y':'rep2_y','Replicate.3_y':'rep3_y','AVERAGE_y': 'AVG_y', 'STDEV_y':'STD_y', 'Free_energy': 'dG', 'functional_x':'is_funct_x', 'functional_y':'is_funct_y'}, inplace = True)
#print Ec_term_df_new.columns
#display(Ec_term_df_new[Ec_term_df_new['Name'].isna() | Ec_term_df_new['term_name'].isna()])

# Delete and reorder columns
proper_cols = ['term_name', 'sequence', 'structure', 'dG', 
 'rep1_x', 'rep2_x', 'rep3_x', 'AVG_x', 'STD_x', 'SEM_x', 'is_funct_x', 'context_x',
 'rep1_y', 'rep2_y', 'rep3_y', 'AVG_y', 'STD_y', 'SEM_y', 'is_funct_y', 'context_y', 
 'term_start', 'term_end', 'term_strand', 
 'up_gene_name',  'up_dist', 'up_gene_start', 'up_gene_end', 'up_gene_strand', 'up_gene_descr',
 'dn_gene_name',  'dn_dist', 'dn_gene_start', 'dn_gene_end', 'dn_gene_strand', 'dn_gene_descr']

Ec_term_df_new = Ec_term_df_new[proper_cols]
display(Ec_term_df_new)

Sp_term_df_new = Sp_term_df_new[proper_cols]
display(Sp_term_df_new)

# Save dataframes in .tsv
Ec_term_df_new.to_csv(Ec_map_updn_term_file_path, sep='\t', index=False, na_rep = 'NA')
Sp_term_df_new.to_csv(Sp_map_updn_term_file_path, sep='\t', index=False, na_rep = 'NA')
````
